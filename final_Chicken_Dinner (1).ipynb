````python
len(tournaments_matches_json["included"])
````

````python
# 해당 경기 참여자 데이터 프레임 변환 방법
sample = [tournaments_matches_json['included'][0]["attributes"]["stats"]]
pd.DataFrame(sample)
````

````python
# 데이터 프레임으로 만들기
participant_indexes = []
for i, item in enumerate(tournaments_matches_json['included']):
    if item.get('type') == 'participant':
        participant_indexes.append(i)

dfs = []
for i in participant_indexes:
    tournaments_matches_player = tournaments_matches_json['included'][i]['attributes']['stats']
    df_tournaments_matches_player = pd.DataFrame(tournaments_matches_player, index=[0])
    dfs.append(df_tournaments_matches_player)

result_df = pd.concat(dfs, ignore_index=True)
result_df
````

### 무기 정보 가져오기

이 코드는 PUBG API를 통해 특정 플레이어의 무기 숙련도 정보를 가져오는데, 

그 중에서도 Item_Weapon_ACE32_C이라는 특정 무기에 대한 정보를 나타냅니다.

해당 무기에 대한 정보는 JSON 형식으로 표현되며, 다음과 같은 정보들을 담고 있습니다


---



* XPTotal: 해당 무기의 총 경험치
* LevelCurrent: 해당 무기의 현재 레벨
* TierCurrent: 해당 무기의 현재 티어
* StatsTotal: 해당 무기를 사용하여 기록한 통계 정보를 담은 객체
* MostDefeatsInAGame: 한 게임에서 가장 많은 사망을 기록한 횟수
* Defeats: 총 사망 횟수
* MostDamagePlayerInAGame: 한 게임에서 가장 많은 데미지를 입힌 플레이어의 데미지 값
* DamagePlayer: 총 데미지
* MostHeadShotsInAGame: 한 게임에서 가장 많은 헤드샷을 기록한 횟수
* HeadShots: 총 헤드샷 횟수
* LongestDefeat: 가장 긴 사망 시간
* LongRangeDefeats: 원거리에서 기록한 사망 횟수
* Kills: 총 킬 횟수
* MostKillsInAGame: 한 게임에서 가장 많은 킬을 기록한 횟수
* Groggies: 총 그로기(상대방을 기절 상태로 만든 것) 횟수
* MostGroggiesInAGame: 한 게임에서 가장 많은 그로기를 기록한 횟수
* OfficialStatsTotal: 해당 무기를 사용하여 기록한 공식 통계 정보를 담은 객체
* MostDefeatsInAGame: 한 게임에서 가장 많은 사망을 기록한 횟수
* Defeats: 총 사망 횟수
* DamagePlayer: 총 데미지
* HeadShots: 총 헤드샷 횟수
* Kills: 총 킬 횟수
* MostKillsInAGame: 한 게임에서 가장 많은 킬을 기록한 횟수
* Groggies: 총 그로기 횟수
* LongestKill: 가장 멀리서 킬한 거리
* CompetitiveStatsTotal: 해당 무기를 사용하여 기록한 경쟁 통계 정보를 담은 객체
* MostDefeatsInAGame: 한 게임에서 가장 많은 적을 물리친 경우의 처치 수
